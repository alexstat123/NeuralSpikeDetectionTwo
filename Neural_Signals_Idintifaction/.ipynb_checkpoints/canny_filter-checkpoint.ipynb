```python
from sklearn.decomposition import PCA
plt.style.use('dark_background')
```

```python
from skimage import color
from skimage import io
X = color.rgb2gray(io.imread('retina3.png'))

plt.figure(figsize=(25,7.5))
plt.gray()
plt.imshow(X)
plt.colorbar()
plt.show()

sigma = 0.5
# load data
X = X.astype(float)
# add noise
Xnoisy = X + sigma*np.random.randn(X.shape[0], X.shape[1])

plt.figure(figsize=(25,7.5))
plt.gray()
plt.imshow(Xnoisy)
plt.colorbar()
plt.show()
```

```python
edges = skimage.feature.canny(
    image=Xnoisy,
    sigma=1,
    low_threshold=0.1,
    high_threshold=0.3,
)
```

```python
plt.figure(figsize=(25,7.5))
plt.gray()
plt.imshow(edges)
plt.colorbar()
plt.show()
```

```python
arr2D = []
for i in numpy.arange(0, 4, 0.1):
    
    edges = skimage.feature.canny(
    image=Xnoisy,
    sigma=i,
    low_threshold=0.1,
    high_threshold=0.3,
)
    edges = edges.flatten()
    edges = edges * 1
    arr2D.append(edges)
    print(i, end=', ')
```

```python
arr2D
```

```python
transposearr2D = np.transpose(arr2D)
```

```python
pca = PCA(n_components = 2)
pca.fit(transposearr2D)
Z = pca.transform(transposearr2D)
Z.shape
```

```python
plt.figure(figsize=(20,20))
plt.scatter(Z[:,0],Z[:,1],s=1)
plt.title('PCA wavelets scatter')
plt.axis('equal')
plt.show()
```
