```python


import numpy as np
from scipy import ndimage as img
from scipy import io as sio
import matplotlib.pyplot as plt
import pyshearlab
import scipy.misc
import cv2
import imageio
from sklearn.decomposition import PCA
from collections import Counter
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
plt.style.use('dark_background')


def shearlets(matrix,sigma,scales,thresholdingFactor):
#     sigma = 1
#     scales =1
    shear = np.array([1, 1, 1])
#     thresholdingFactor = 200
    # load data
#     X = imageio.imread('barbara.jpg')[::4, ::4]
    matrix = matrix.astype(float)
    x = matrix 
    # add noise
#     Xnoisy = X + sigma*np.random.randn(X.shape[0], X.shape[1])
    ## create shearlets
    shearletSystem = pyshearlab.SLgetShearletSystem2D(0,x.shape[0], x.shape[1], scales)
#     shearletSystem = pyshearlab.SLgetShearletSystem2D(0,X.shape[0], X.shape[1], scales,shear,1)
    # decomposition
    coeffs = pyshearlab.SLsheardec2D(x, shearletSystem)
    # thresholding
    oldCoeffs = coeffs.copy()
    weights = np.ones(coeffs.shape)

    for j in range(len(shearletSystem["RMS"])):
        weights[:,:,j] = shearletSystem["RMS"][j]*np.ones((x.shape[0], x.shape[1]))

    coeffs = np.real(coeffs)
    zero_indices = np.abs(coeffs) / (thresholdingFactor * weights * sigma) < 1
    coeffs[zero_indices] = 0
    
    print("coeffs shape", coeffs.shape)
    
#   ////////////////// my func ////////////////////////////////////////////////
#     v = np.vstack(coeffs)
#     pca = PCA(n_components = 1)
#     pca.fit(v)
#     Z = pca.transform(v)
    
#     kmeans = KMeans(n_clusters=1).fit(Z)
#     centroids = kmeans.cluster_centers_
#     Distance = cdist(Z, centroids)
#     Distance
    
#     satandardDiv = np.std(Z)
#     threshold = 3 * satandardDiv
    
#     mask = map(lambda x : 1 if (x > threshold) else 0, Distance)
#     mask = list(mask)
    
#     m = np.array(v)
#     c = np.array(mask)
    
#     V = m * c[:, np.newaxis]
#     V = np.array(V).reshape(128,128,17)
    
#  ///////////////////////////////////////////////////////////////////////////////

    # reconstruction
    Xrec = pyshearlab.SLshearrec2D(coeffs, shearletSystem)

    PSNR = pyshearlab.SLcomputePSNR(x,Xrec)
    print("PSNR: " + str(PSNR))
    return Xrec, coeffs, scales, weights
```

```python
X = imageio.imread('barbara.jpg')[::4, ::4]
# M = np.random.rand(128,128)+100
# NoisyImage = X + M*100
sigma = 100
Xnoisy = X + sigma*np.random.randn(X.shape[0], X.shape[1])

# print(X.shape)
Xrec, coeffs,scales,weights = shearlets(matrix = Xnoisy, sigma = 1,scales = 2,thresholdingFactor = 200)


plt.figure(figsize=(25,7.5))
plt.gray()
# plt.imshow(Xrec) 
plt.imshow(X) 
plt.colorbar()
plt.show()

plt.figure(figsize=(25,7.5))
plt.gray()
plt.imshow(Xnoisy) 
plt.colorbar()
plt.show()

plt.figure(figsize=(25,7.5))
plt.gray()
plt.imshow(Xrec) 
plt.colorbar()
plt.show()
    
```

```python
v = np.vstack(coeffs)
# v = np.vstack(weights)
```

```python
v.shape
```

```python
pca = PCA(n_components = 1)
pca.fit(v)
Z = pca.transform(v)
Z.shape
```

```python
n, bins, patches = plt.hist(Z,bins=200)
plt.show()
```

```python
satandardDiv = np.std(Z)
threshold = 1 * satandardDiv
threshold
```

```python
kmeans = KMeans(n_clusters=1).fit(Z)
centroids = kmeans.cluster_centers_
Distance = cdist(Z, centroids)
Distance
```

```python
mask = map(lambda x : 1 if (x > threshold) else 0, Distance)
mask = list(mask)
mask = np.array(mask).reshape(128,128)
rec = Xnoisy*mask
plt.figure(figsize=(25,7.5))
plt.gray()
plt.imshow(rec)
plt.colorbar()
plt.show()
```

```python
mask = map(lambda x : 1 if (x < -100) else 0, Z)
mask = list(mask)
mask = np.array(mask).reshape(128,128)
rec = Xnoisy*mask
plt.figure(figsize=(25,7.5))
plt.gray()
plt.imshow(rec)
plt.colorbar()
plt.show()
```

```python
pca = PCA(n_components = 2)
pca.fit(v)
Z = pca.transform(v)
# plot PCA
plt.figure(figsize=(20,20))
plt.scatter(Z[:,0],Z[:,1],s=1)
plt.title('PCA wavelets scatter')
plt.axis('equal')
plt.show()
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
# from sklearn.datasets import make_blobs
bandwidth = estimate_bandwidth(Z, quantile=0.1, n_samples=150)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(Z)
labels = ms.labels_
cluster_centers = ms.cluster_centers_
labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)
print("number of estimated clusters : %d" % n_clusters_)

from itertools import cycle
plt.figure(1)
plt.clf()
colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
plt.figure(figsize=(10,10))
for k, col in zip(range(n_clusters_), colors):
    my_members = labels == k
    cluster_center = cluster_centers[k]
    plt.plot(Z[my_members, 0], Z[my_members, 1], col + '.')
    plt.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=14)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()


```
